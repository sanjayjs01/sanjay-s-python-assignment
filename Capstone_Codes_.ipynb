```python
df['order_time']=df['order_time'].astype('string')
df[['Hour','Minute', 'Second']]= df['order_time'].str.split(":",expand=True)
```

```python
df['datetime'] = pd.to_datetime(df['order_date'].astype(str) + ' ' + df['order_time'].astype(str))
df['weekday'] = df['datetime'].dt.strftime('%A')
```

Hypothesis 01:
Finding the Daywise Trend in Pizza Sale

```python
# Group the data by weekday and count the number of unique order IDs for each weekday
weekday_orders = df.groupby('weekday')['order_id'].nunique()


# Define the order of weekdays
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


# Reindex the weekday_orders series to match the weekday_order
weekday_orders = weekday_orders.reindex(weekday_order)


# Plot the bar chart

plt.bar(weekday_orders.index, weekday_orders.values)
plt.xlabel('Weekday',color = 'blue')
plt.ylabel('Number of Orders')
plt.title('Orders per Weekday',fontsize = 20,color = 'green')


plt.xticks(rotation=45)  # Tilt the x-axis labels by 45 degrees
plt.show()
```

```python
# Group the data by weekday and count the number of unique order IDs for each weekday
weekday_orders = df.groupby('weekday')['order_id'].nunique()


# Print the number of orders for each weekday
print("Orders per Weekday:")
for weekday, count in weekday_orders.items():
    print(f"{weekday}: {count} orders")
```

Hypothesis 02
What is the busiest time in the day?

```python
sns.countplot(data=df,x="Hour",palette="copper")
plt.xticks(rotation=45)
plt.xlabel("Hour",fontsize=10,color="blue")
plt.ylabel("Frequency",fontsize=10,color="purple")
plt.title("HOUR",color="purple")
plt.show()
```

```python
# Calculate the number of unique order_id per hour

orders_per_hour = df.groupby('Hour')['order_id'].nunique()


# Create the line chart

plt.plot(orders_per_hour.index, orders_per_hour.values)
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.title('Peak Hour in a Day(24 hours format)',color ='green')
plt.grid()
plt.show()
```

```python
# Group the data by hour and count the number of unique orders for each hour

hourly_orders = df.groupby(df['datetime'].dt.hour)['order_id'].nunique()


# Print the count of unique orders for each hour

for hour, count in hourly_orders.items():
    print(f"Hour {hour}: {count} orders")
```

Hupothesis 03 :
Best Selling Pizza Category ?


```python
pizza_count = df['pizza_category'].value_counts()
best_selling_pizza = pizza_count.idxmax()
worst_selling_pizza = pizza_count.idxmin()
print("Best Selling Pizza:",best_selling_pizza)
print("Worst Selling Pizza:",worst_selling_pizza)
```

```python
labels = df["pizza_category"].value_counts().index
sizes = df["pizza_category"].value_counts()
colors = ['#ff9999','#66b3ff','#990f99','#ffcc89',"pink","yellow"]
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors,shadow=True, startangle=45)
plt.show()
```

Hypothesis 04 :
How many orders are we expecting each day?

```python
# Calculate the number of customers per day
customers_per_day = df.groupby(df['order_date'])['order_id'].nunique().mean()

print("Number of orders we are expecting per day:")
print(customers_per_day)
```

Hypothesis 04 :
Revenue Per Day


```python
df.groupby('order_date').agg({'total_price': 'sum'}).plot(kind = 'line', figsize=(15,7), legend=False, color="blue")
plt.ylabel('Revenue')
plt.title('Revenue per day');
```

Monthly Sales Trends?


```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['order_month'] = df['order_date'].dt.month
monthly_order_count = df.groupby('order_month').size()
plt.figure(figsize=(10, 5))
monthly_order_count.plot(kind='line', marker='o')
plt.xlabel("Month")
plt.ylabel("Total sales")
plt.title("Monthly Sales Trend")
plt.xticks(range(1, 13), ['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'] )
plt.grid(True)
plt.show()
```
